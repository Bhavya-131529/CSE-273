```python
import pandas as pd

df = pd.DataFrame({
    'EmployeeName': ['Alice', 'Bob', 'Charlie', 'David', 'Eve'],
    'Department': ['HR', 'Engineering', 'Marketing', 'Sales', 'Material Dep'],
    'Role': ['Analyst', 'Manager', 'Employee', 'Senior Eng', 'Senior Eng'],
    'Salary': [60000, 80000, 75000, 65000, 90000],
    'Experience': [2, 5, 3, 4, 8],
    'Score':[95,75,68,95,89]
})
print(df)
```

```python
average= df.groupby('Department')['Salary'].mean()
print(average)
```

```python
avg_score_by_role = df.groupby('Role')['Score'].mean()
print(avg_score_by_role)
```

```python
non_null = df.groupby('Department')['Score'].count()
print(non_null)
```

```python
total_records_by_department = df.groupby('Department').size()
print(total_records_by_department)
```

```python
summary= df.groupby('Department').agg(average_salary=('Salary', 'mean'), non_null=('Score', 'count'), total=('EmployeeName', 'size'))
print('Departmental Summary:')
print(summary)
```
